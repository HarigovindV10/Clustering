```python
import math
def euclidean_distance(x,y):
    s=0
    for i,j in zip(x,y):
        s+=(i-j)**2
    return math.sqrt(s)
```

```python
def minimum(a):
    temp=list(a)
    a.sort()
    b=a[0]
    return temp.index(b)
```

```python
import random
ri=[]
k1=[]
k2=[]
k3=[]

k=3
ri=random.sample(range(0,len(df)),3)
#ri=[25,50,75]
for i in range(len(df)):
    
    m=[]
    m.append(euclidean_distance(df['re'][ri[0]],df['re'][i]))
    m.append(euclidean_distance(df['re'][ri[1]],df['re'][i]))
    m.append(euclidean_distance(df['re'][ri[2]],df['re'][i]))
    c=minimum(m)
    #print(c)
    if c==0:
        k1.append(df['re'][i])
    elif c==1:
        k2.append(df['re'][i])
    elif c==2:
        k3.append(df['re'][i])

#print(k2)
#print(type(k2))
#print(type(k2[0]))
for j in range(10):
    #ri=[]
    s1,s2=0,0
    #print(len(k1),len(k2),len(k3))
    print(ri)
    x1,y1=zip(*k1)
    #print(type(x1),len(x1),x1)
    for x2 in x1:
        
        s1+=x2
        
    s1/=len(k1)
    for x2 in y1:
        s2+=x2
    s2/=len(k1)
    ri[0]=(s1,s2)
    s1,s2=0,0
    x1,y1=zip(*k2)
    for x2 in x1:
        s1+=x2
        
    s1/=len(k2)
    for x2 in y1:
        s2+=x2
    s2/=len(k2)
    ri[1]=(s1,s2)
    s1,s2=0,0
    x1,y1=zip(*k3)
    for x2 in x1:
        
        s1+=x2
        
    s1/=len(k3)
    for x2 in y1:
        s2+=x2
    s2/=len(k3)
    ri[2]=(s1,s2)
    k1,k2,k3=[],[],[]
    for i in range(len(df)):
        m=[]
        m.append(euclidean_distance(ri[0],df['re'][i]))
        m.append(euclidean_distance(ri[1],df['re'][i]))
        m.append(euclidean_distance(ri[2],df['re'][i]))
        c=minimum(m)
        #print(c)
        if c==0:
            k1.append(df['re'][i])
        elif c==1:
            k2.append(df['re'][i])
        elif c==2:
            k3.append(df['re'][i])
print("\n1st cluster\n")
for i in k1:
    print(i)
print("\n2nd cluster\n")
for i in k2:
    print(i)
print("\n3rd cluster\n")
for i in k3:
    print(i)
```
